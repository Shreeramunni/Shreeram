## Multiple Faces

```
!wget https://cdn.cnn.com/cnnnext/dam/assets/170305143551-trump-obama-split-exlarge-169.jpg -O multiple_unknown.jpg
```

```
unknown_multiple = face_recognition.load_image_file('multiple_unknown.jpg')
face_locations = face_recognition.face_locations(unknown_multiple)
print(face_locations)
```

```
face_encodings = face_recognition.face_encodings(unknown_multiple, face_locations)
print(face_encodings)
```

```
unknown_bgr = cv2.cvtColor(unknown_multiple, cv2.COLOR_RGB2BGR)
for face, face_encoding in zip(face_locations, face_encodings):
  (top, right, bottom, left) = face
  cv2.rectangle(unknown_bgr, (left,top), (right,bottom), (0,0,255), 2)
  result = face_recognition.compare_faces([known_encoding], face_encoding)
  dist = face_recognition.face_distance([known_encoding], face_encoding)
  print(dist)
  '''print(result)
  if result[0] == True:
    cv2.putText(unknown_bgr, "Obama", (left,top - 10), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0,255,0), 1)
  if result[0] == False:
    cv2.putText(unknown_bgr, "Not Obama", (left,top - 10), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0,0,255), 1)'''

  if dist[0] < 0.5:
    cv2.putText(unknown_bgr, "Obama", (left,top - 10), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0,255,0), 1)
  else:
    cv2.putText(unknown_bgr, "Not Obama", (left,top - 10), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0,0,255), 1)
```
