## Image Classification with Keras

```
# importing necessary packages
import keras
import cv2
from keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
from keras_preprocessing import image
import numpy as np 
from google.colab.patches import cv2_imshow
```

```
# load pre-trained CNN model
model = ResNet50(weights="imagenet")
```

```
!wget https://upload.wikimedia.org/wikipedia/commons/4/45/Cherry_fruit_on_white_background.jpg
```

```
img= image.load_img('Cherry_fruit_on_white_background.jpg',target_size=(224,224))
```

```
img= image.img_to_array(img)
img= np.expand_dims(img,axis=0)
img= preprocess_input(img)
```

```
preds= model.predict(img)
print(preds)
```

```
label =decode_predictions(preds, top=5)[0]
print(label)
label =str(label[0][1])+":"+str(label[0][2])
print(label)
```

```
out =cv2.imread('Cherry_fruit_on_white_background.jpg')
cv2.putText(out,label,(50,50),cv2.FONT_HERSHEY_SIMPLEX,1.0,(0,255,0),2)
```

```
cv2_imshow(out)
```
